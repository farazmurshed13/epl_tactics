```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
epl_1920 = pd.read_csv('epl_19_20.csv')
#epl_1920.head()
```

```python
# goal stats vs points
plt.scatter(epl_1920['GF'], epl_1920['Pts'], label='Goals For')
plt.scatter(epl_1920['GA'], epl_1920['Pts'], label='Goals Allowed')
plt.xlabel('Goals')
plt.ylabel('Points')
plt.title('EPL Goal Statistics vs. Points Won')
plt.legend()
plt.show()

# passing vs points
plt.scatter(epl_1920['Pass_perc'], epl_1920['Pts'])
plt.xlabel('Passing %')
plt.ylabel('Points')
plt.title('Passing % vs. Points Won')
plt.show()
```

```python
# offensive metrics

# possession/passing vs GF
plt.scatter(epl_1920['Possession_perc'], epl_1920['GF'], label='Possession %')
plt.scatter(epl_1920['Pass_perc'], epl_1920['GF'], label='Pass %')
plt.xlabel('%')
plt.ylabel('Goals For')
plt.title('Possession Statistics vs. Goals For')
plt.legend()
plt.show()

# shooting vs GF
plt.scatter(epl_1920['Shots_pg'], epl_1920['GF'], label='Shots per game')
plt.scatter(epl_1920['Shots_OT_pg'], epl_1920['GF'], label='Shots on-target per game')
plt.xlabel('Shots')
plt.ylabel('Goals For')
plt.title('Shooting Statistics vs. Goals For')
plt.legend()
plt.show()

# dribbling vs GF
plt.scatter(epl_1920['Dribbles_pg'], epl_1920['GF'])
plt.xlabel('Dribbles per game')
plt.ylabel('Goals For')
plt.title('Dribbling Statistics vs. Goals For')
plt.show()

# possession vs passing
plt.scatter(epl_1920['Possession_perc'], epl_1920['Pass_perc'])
plt.xlabel('Possession %')
plt.ylabel('Passing %')
plt.title('Possession % vs. Passing %')
plt.show()

# shooting vs possession
plt.scatter(epl_1920['Shots_pg'], epl_1920['Possession_perc'])
plt.xlabel('Shots per Game')
plt.ylabel('Possession %')
plt.title('Shots per Game % vs. Possession %')
plt.show()

# dribbling vs possession
plt.scatter(epl_1920['Dribbles_pg'], epl_1920['Possession_perc'], color='purple')
plt.xlabel('Dribbles per Game')
plt.ylabel('Possession %')
plt.title('Dribbling Statistics vs. Possession %')
plt.show()

# dribbling vs shooting
plt.scatter(epl_1920['Dribbles_pg'], epl_1920['Shots_pg'], color='r')
plt.xlabel('Dribbles per Game')
plt.ylabel('Shots per Game')
plt.title('Dribbling Statistics vs. Shots per Game')
plt.show()
```

```python
# defensive metrics

# shots allowed vs GA
plt.scatter(epl_1920['Shots_allowed_pg'], epl_1920['GA'])
plt.xlabel('Shots Allowed per game')
plt.ylabel('Goals Allowed')
plt.title('Shots Allowed vs. Goals Allowed')
plt.show()

# tackles vs GA
plt.scatter(epl_1920['Tackles_pg'], epl_1920['GA'])
plt.xlabel('Tackles per game')
plt.ylabel('Goals Allowed')
plt.title('Tackles vs. Goals Allowed')
plt.show()

# interceptions vs GA
plt.scatter(epl_1920['Interceptions_pg'], epl_1920['GA'])
plt.xlabel('Interceptions per game')
plt.ylabel('Goals Allowed')
plt.title('Interceptions vs. Goals Allowed')
plt.show()

# interceptions vs Shots Allowed
plt.scatter(epl_1920['Interceptions_pg'], epl_1920['Shots_allowed_pg'])
plt.xlabel('Interceptions per game')
plt.ylabel('Shots Allowed')
plt.title('Interceptions vs. Shots Allowed')
plt.show()

# tackles vs Shots Allowed
plt.scatter(epl_1920['Tackles_pg'], epl_1920['Shots_allowed_pg'])
plt.xlabel('Tackles per game')
plt.ylabel('Shots Allowed')
plt.title('Tackles vs. Shots Allowed')
plt.show()

# possession vs Shots Allowed
plt.scatter(epl_1920['Possession_perc'], epl_1920['Shots_allowed_pg'])
plt.xlabel('Possession %')
plt.ylabel('Shots Allowed')
plt.title('Possession % vs. Shots Allowed')
plt.show()
```
